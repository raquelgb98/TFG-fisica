```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from qutip import *
```

$\displaystyle H = - \frac{1}{2}\lambda\sum_n^N  \sigma_z(n) - \frac{1}{2} \sum_n^{N-1} [ \frac{1+\gamma}{2}\sigma_x(n) \sigma_x(n+1) + \frac{1-\gamma}{2}\sigma_y(n) \sigma_y(n+1)]$

```python
def integrate(N, l, g, psi0, tlist, gamma, solver = 'me', periodic = False):
    """N - length of the spin-chain
    l - lambda 
    g - gamma
    psi0 - initial state
    tlist - times for the evolution
    gamma - parameters of collapse operators
    solver - type of qutip solver
    periodic - True for periodic boundary conditions, False for open boundary conditions"""
    
    si = qeye(2)
    sx = sigmax()
    sy = sigmay()
    sz = sigmaz()

    sx_list = []
    sy_list = []
    sz_list = []

    for n in range(N):
        op_list = []
        for m in range(N):
            op_list.append(si)

        op_list[n] = sx
        sx_list.append(tensor(op_list))

        op_list[n] = sy
        sy_list.append(tensor(op_list))

        op_list[n] = sz
        sz_list.append(tensor(op_list))

    # construct the hamiltonian
    H = 0

    # energy splitting terms
    for n in range(N):
        H += - 0.5 * l * sz_list[n]

    # interaction terms
    for n in range(N-1):
        H += - 0.5 * (1+g) * 0.5 * sx_list[n] * sx_list[n+1]
        H += - 0.5 * (1-g) * 0.5 * sy_list[n] * sy_list[n+1]

    # adding terms for periodic boundary condition, interaction of spin N and spin 1
    if periodic == True:
        H += - 0.5 * (1+g) * 0.5 * sx_list[N-1] * sx_list[0]
        H += - 0.5 * (1-g) * 0.5 * sy_list[N-1] * sy_list[0]
        
    # collapse operators
    c_op_list = []

    # spin dephasing
    for n in range(N):
        if gamma[n] > 0.0:
            c_op_list.append(np.sqrt(gamma[n]) * sz_list[n])

    # evolve and calculate expectation values
    if solver == "me":
        result = mesolve(H, psi0, tlist, c_op_list, sz_list)
        result_state = mesolve(H, psi0, tlist, c_op_list)
    elif solver == "mc":
        ntraj = 250 
        result = mcsolve(H, psi0, tlist, c_op_list, sz_list, ntraj)

    return result.expect, result_state.states

def entanglement_entropy(state, partition):
    # Returns the Von Neumann entropy of the partial trace
    # state - state for which we want to calculate the entanglement entropy
    # partition - integer or array of integers from 0 to N-1 that indicate the subsystem
    # with respect to which we calculate the partial trace
    dm = state*state.dag()
    partial_dm = dm.ptrace(partition)
    return entropy_vn(partial_dm, base = 2)

def generate_state(N, llista):
    # N - total number of spins
    # llista - list of numbers from 0 to N-1 that indicate when a spin is flipped
    a = np.zeros(N)
    for i in llista:
        a[i] = 1
    basis_list = []
    for i in range(N):
        if a[i] == 0:
            basis_list.append(basis(2,0))
        if a[i] == 1:
            basis_list.append(basis(2,1))
    state = tensor(basis_list)
    return state
```

```python
#
# set up the calculation
#
solver = "me"   # use the ode solver
#solver = "mc"   # use the monte-carlo solver

N = 12            # number of spins


l = 2
g = 1
gamma = np.zeros(N) #0.01 * np.ones(N)

# initial state is a superposition of two states with spins flipped at positions 5 and 6 respectively
# initial state is entangled
state1 = generate_state(N, [5])
state2 = generate_state(N, [6])
psi0 = (state1+state2).unit()

tmin = 0
tmax = 50
tlist = np.linspace(tmin, tmax, 200)

expt, states = integrate(N, l, g, psi0, tlist, gamma, solver, periodic = True)
```

```python
plt.subplots(figsize=(10,6))
plt.imshow(np.real(expt), interpolation = 'nearest', extent=[tmin, tmax, 0,N ], aspect = 'auto')
plt.aspect_ratio = 1
plt.colorbar()
plt.xlabel(r'Time [ns]')
plt.ylabel(r'N site')
plt.title(r'$\langle\sigma_z\rangle$ evolution of a spin chain');
```

```python
entropy = np.zeros((N, len(tlist)))
for i in range(N):
    j = 0
    for state in states:
        entropy[i,j] = entanglement_entropy(state, i)
        j+= 1
```

```python
plt.subplots(figsize = (10,6))
for i in range(N):
    plt.plot(tlist, entropy[i], label = r'%d'%i)
plt.xlabel('Time')
plt.ylabel('Entanglement entropy')
plt.legend(loc = 1)
```

```python
plt.subplots(figsize=(10,6))
plt.imshow(entropy, interpolation = 'nearest', extent=[tmin, tmax, 0,N ], aspect = 'auto')
plt.colorbar()
plt.xlabel(r'Time [ns]')
plt.ylabel(r'N site')
plt.title(r'Entanglement entropy evolution of a spin chain');
```

```python
total_entropy = entropy.sum(axis = 0)
plt.subplots(figsize = (10,6))
plt.plot(tlist, total_entropy)
plt.xlabel('Time')
plt.ylabel('Sum of entanglement entropies')
```

```python
N = 10
l = 2
g = 1

# Initial state has a spin flipped at position 4
# Initial state is not entangled
psi0= generate_state(N, [4])
gamma = np.zeros(N)
tmin = 0
tmax = 50
tlist = np.linspace(tmin, tmax, 200)

expt, states = integrate(N, l, g, psi0, tlist, gamma, periodic = False)
expt_p, states_p = integrate(N, l, g, psi0, tlist, gamma, periodic = True)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10,6))
fig.suptitle(r'$\langle\sigma_z\rangle$ evolution of a spin chain')
a = ax1.imshow(np.real(expt), interpolation = 'nearest', extent=[tmin, tmax, 0,N ], 
               aspect = 'auto', label = 'Open chain')
ax1.set_ylabel(r'N site')
ax1.set_title('Periodic boundary condition')
plt.colorbar(a, ax = ax1)

b = ax2.imshow(np.real(expt_p), interpolation = 'nearest', extent=[tmin, tmax, 0,N ],
               aspect = 'auto', label = 'Periodic chain')
plt.colorbar(b, ax = ax2)
ax2.set_title('Open boundary condition')
ax2.set_xlabel(r'Time [ns]')
ax2.set_ylabel(r'N site')
```

```python
entropy = np.zeros((N, len(tlist)))
for i in range(N):
    j = 0
    for state in states:
        entropy[i,j] = entanglement_entropy(state, i)
        j+= 1
        
entropy_p = np.zeros((N, len(tlist)))
for i in range(N):
    j = 0
    for state_p in states_p:
        entropy_p[i,j] = entanglement_entropy(state_p, i)
        j+= 1
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10,6))
fig.suptitle(r'Entanglement entropy evolution of a spin chain')
a = ax1.imshow(entropy, interpolation = 'nearest', extent=[tmin, tmax, 0,N ], 
               aspect = 'auto', label = 'Open chain')
ax1.set_ylabel(r'N site')
ax1.set_title('Periodic boundary condition')
plt.colorbar(a, ax = ax1)

b = ax2.imshow(entropy_p, interpolation = 'nearest', extent=[tmin, tmax, 0,N ],
               aspect = 'auto', label = 'Periodic chain')
plt.colorbar(b, ax = ax2)
ax2.set_title('Open boundary condition')
ax2.set_xlabel(r'Time [ns]')
ax2.set_ylabel(r'N site')
```

```python
total_entropy = entropy.sum(axis = 0)
total_entropy_p = entropy_p.sum(axis = 0)
plt.subplots(figsize = (10,6))
plt.plot(tlist, total_entropy, label ='OBC')
plt.plot(tlist, total_entropy_p, label = 'PBC')
plt.xlabel('Time')
plt.ylabel('Sum of entanglement entropies')
plt.legend(loc = 0)
```
