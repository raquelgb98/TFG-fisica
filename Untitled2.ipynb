```python
from openfermion.chem import MolecularData
from openfermionpyscf import run_pyscf
#from openfermion.transforms import get_fermion_operator, jordan_wigner

geometry = [['H',[0,0,0]],
           ['H',[0,0,0.74]]] #0.74
basis = 'sto-3g'
multiplicity = 1
charge = 0
h2_molecule = MolecularData(geometry, basis, multiplicity, charge)
#h2_molecule = run_pyscf(h2_molecule,
#                       run_mp2 = True,
 #                      run_cisd = True,
  #                     run_ccsd = True,
   #                    run_fci = True)
h2_molecule.load()
#h2_filename = h2_molecule.filename
#h2_molecule.save()
h2_hamiltonian = h2_molecule.get_molecular_hamiltonian()

#fermion_hamiltonian = get_fermion_operator(h2_hamiltonian)
#qubit_hamiltonian = jordan_wigner(fermion_hamiltonian)
```

```python
from openfermion.circuits import simulate_trotter
from openfermion.circuits.trotter import LOW_RANK
import openfermion
import cirq
import qutip as qp
# Initialize a random initial state.
import numpy as np

n_qubits = openfermion.count_qubits(h2_hamiltonian)
random_seed = 8317
initial_state = openfermion.haar_random_vector(
    2 ** n_qubits, random_seed).astype(np.complex64)
# Numerically compute the correct circuit output.

# Trotter step parameters.
n_steps = 3
time = np.arange(0,40,0.1)
entropy1 = []
entropy2 = []
entropy3 = []
entropy4 = []
entropy5 = []
entropy6 = []
# Compile the low rank Trotter step using OpenFermion.
qubits = cirq.LineQubit.range(n_qubits)
for t in time:
    circuit = cirq.Circuit(
        simulate_trotter(
            qubits, h2_hamiltonian,
            time=t, n_steps=n_steps,
            algorithm=LOW_RANK),
    strategy=cirq.InsertStrategy.EARLIEST)

# Use Cirq simulator to apply circuit.
    simulator = cirq.DensityMatrixSimulator()
    result = simulator.simulate(circuit, initial_state=initial_state)
#result = simulator.simulate(circuit, qubit_order=qubits, initial_state=initial_state)
    step_matrix = result.final_density_matrix
    step_dm = qp.Qobj(step_matrix,dims = [[2,2,2,2],[2,2,2,2]])
    step_partial_trace = step_dm.ptrace([0,1])
    entropy1.append(cirq.von_neumann_entropy(step_partial_trace))
    step_partial_trace = step_dm.ptrace([0,2])
    entropy2.append(cirq.von_neumann_entropy(step_partial_trace))
    step_partial_trace = step_dm.ptrace([0,3])
    entropy3.append(cirq.von_neumann_entropy(step_partial_trace))
    step_partial_trace = step_dm.ptrace([1,2])
    entropy4.append(cirq.von_neumann_entropy(step_partial_trace))
    step_partial_trace = step_dm.ptrace([1,3])
    entropy5.append(cirq.von_neumann_entropy(step_partial_trace))
    step_partial_trace = step_dm.ptrace([2,3])
    entropy6.append(cirq.von_neumann_entropy(step_partial_trace))
```

```python
import matplotlib.pyplot as plt
plt.plot(time, entropy1, label = '[0,1]')
plt.plot(time, entropy2, label = '[0,2]')
plt.plot(time, entropy3, label = '[0,3]')
entropy_list = [entropy1, entropy2, entropy3]
entropy = np.mean(entropy_list, axis = 0)
plt.plot(time, entropy, label = 'mean')
#plt.plot(time, entropy4, label = '[1,2]')
#plt.plot(time, entropy5, label = '[1,3]')
#plt.plot(time, entropy6, label = '[2,3]')
plt.xlabel('time')
plt.ylabel('Entanglement entropy')
plt.legend()
```

```python
from openfermion.circuits import simulate_trotter
from openfermion.circuits.trotter import LOW_RANK
import openfermion
import cirq
import qutip as qp
import numpy as np

def simulate_entanglement_entropy(hamiltonian, initial_state,time, partitions,n_steps = 3):
    n_qubits = openfermion.count_qubits(hamiltonian)
    print('The number of qubits is:', n_qubits)
    initial_state = initial_state
    # Numerically compute the correct circuit output.
    # Trotter step parameters.
    n_steps = n_steps
    time = time
    entropy = np.zeros((len(partitions),len(time)))
    # Compile the low rank Trotter step using OpenFermion.
    qubits = cirq.LineQubit.range(n_qubits)
    for t in range(0,len(time)):
        circuit = cirq.Circuit(
        simulate_trotter(
            qubits, hamiltonian,
            time=time[t], n_steps=n_steps,
            algorithm=LOW_RANK),
    strategy=cirq.InsertStrategy.EARLIEST)

        # Use Cirq simulator to apply circuit.
        simulator = cirq.DensityMatrixSimulator()
        result = simulator.simulate(circuit, initial_state=initial_state)
        #result = simulator.simulate(circuit, qubit_order=qubits, initial_state=initial_state)
        step_matrix = result.final_density_matrix
        step_dm = qp.Qobj(step_matrix,dims = [[2,2,2,2],[2,2,2,2]])
        for p in range(0,len(partitions)): 
            step_partial_trace = step_dm.ptrace(partitions[p])
            entropy[p,t]=cirq.von_neumann_entropy(step_partial_trace)
            
    #plot the results
    import matplotlib.pyplot as plt
    for p in range(0,len(partitions)):
        plt.plot(time, entropy[p], label = str(partitions[p]))
        plt.legend()
        plt.xlabel('time')
        plt.ylabel('Entanglement entropy')

    return entropy
```

```python
partitions = [[0,1],[0,2],[0,3]]
time = np.arange(0,100,0.5)
entropy = simulate_entanglement_entropy(LiH_hamiltonian,initial_state,time,partitions,n_steps = 3)
```
