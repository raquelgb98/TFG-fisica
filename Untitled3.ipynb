```python
from openfermion.circuits import simulate_trotter
from openfermion.circuits.trotter import LOW_RANK
import openfermion
import cirq
from openfermion.transforms import get_fermion_operator
from openfermion import FermionOperator
from openfermion.linalg import sparse_tools
import qutip as qp
import numpy as np
from scipy.linalg import eigh
import scipy

def simulate_entanglement_entropy(hamiltonian, initial_state, time, partitions, exact = True, n_steps = 3):
    initial_state = initial_state
    time = time
    # Numerically compute the correct circuit output.
    # Trotter step parameters.
    
    entropy = np.zeros((len(partitions),len(time)))
    # Compile the low rank Trotter step using OpenFermion.
    # We will a cirq simulator to simulate the time evolution with a quantum circuit
    if exact == False:
        n_qubits = openfermion.count_qubits(hamiltonian)
        print('The number of qubits is:', n_qubits)
        n_steps = n_steps
        qubits = cirq.LineQubit.range(n_qubits)
        for t in range(0,len(time)):
            circuit = cirq.Circuit(
                simulate_trotter(
                qubits, hamiltonian,
                time=time[t], n_steps=n_steps,
                algorithm=LOW_RANK),
                strategy=cirq.InsertStrategy.EARLIEST)

        # Use Cirq simulator to apply circuit.
            simulator = cirq.Simulator()
            result = simulator.simulate(circuit, qubit_order=qubits, initial_state=initial_state)
            simulated_state = result.final_state_vector
            
            fermion_hamiltonian = get_fermion_operator(hamiltonian)
            sparse_operator = sparse_tools.get_sparse_operator(fermion_hamiltonian).todense()
            exact_state = scipy.sparse.linalg.expm_multiply(
                -1j * time[t] * np.array(sparse_operator), initial_state)
            print('fidelity: ',abs(np.dot(exact_state, np.conjugate(simulated_state))) ** 2)
            for p in range(0,len(partitions)): 
                entropy[p][t] = get_entanglement_entropy(simulated_state, partitions[p])
    if exact == True:
        fermion_hamiltonian = get_fermion_operator(hamiltonian)
        sparse_operator = sparse_tools.get_sparse_operator(fermion_hamiltonian).todense()
        for t in range(0,len(time)):
            exact_state = scipy.sparse.linalg.expm_multiply(
                -1j * time[t] * np.array(sparse_operator), initial_state)
            for p in range(0,len(partitions)): 
                entropy[p][t] = get_entanglement_entropy(exact_state, partitions[p])
    #plot the results
    import matplotlib.pyplot as plt
    for p in range(0,len(partitions)):
        plt.plot(time, entropy[p], label = str(partitions[p]))
        plt.legend()
        plt.xlabel('time')
        plt.ylabel('Entanglement entropy')

    return entropy

def get_eigenstates(hamiltonian):
    fermion_hamiltonian = get_fermion_operator(hamiltonian)
    sparse_operator = sparse_tools.get_sparse_operator(fermion_hamiltonian).todense()
    energies, states = eigh(sparse_operator)
    return energies, states

def get_entanglement_entropy(state, partition):
    #start calculating the density matrix
    quantum_state = qp.Qobj(state)
    step_matrix = quantum_state*quantum_state.dag()
    step_dm = qp.Qobj(step_matrix,dims = [[2,2,2,2],[2,2,2,2]])
    step_partial_trace = step_dm.ptrace(partition)
    entropy = cirq.von_neumann_entropy(step_partial_trace)
    return entropy        
```
